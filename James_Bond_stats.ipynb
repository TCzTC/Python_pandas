```python
bond.Shooting_Loc
```

```python
from collections import Counter
all_locations = ", ".join(bond["Shooting_Loc"]).split(", ")
location_counts = Counter(all_locations)
location_counts

top_locations = dict(location_counts)
top_locations

bond_locations = pd.Series(top_locations).head(10).sort_values(ascending=False)
bond_locations
```

```python
bond_locations.plot(kind="bar",figsize=(6,4))
plt.title="Bond Locations"
plt.xlabel="Locations"
plt.ylabel="Count"
plt.show()
```

```python
import plotly.express as px
df_plot = bond_locations.reset_index()
df_plot.columns = ["Location", "Count"]

fig = px.bar(df_plot,
             x="Location",
             y="Count",
             title="Movie locations",
             labels={"Location":"Locations","Count":"Count"})
fig.show()
```
