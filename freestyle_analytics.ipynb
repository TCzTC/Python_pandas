```python
bond["Bond"].apply(len)

def rank_movie(row):
  year = row.loc["Year"]
  bond_actor = row.loc["Bond"]

  if year >= 1990 and year <= 1999:
    return "Best Bonds"

  if bond_actor == "Daniel Craig":
    return "Best Actor"

  return "no description"

bond.apply(rank_movie, axis="columns")
```
